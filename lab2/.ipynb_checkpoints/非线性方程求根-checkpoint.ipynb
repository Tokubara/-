阻尼牛顿的实现见dampedNewton.m，既可以选择阻尼也可以选择无阻尼，默认有阻尼，设定的阻尼因子一开始都是1(需要在0到1之间)，判定准则是：f(x)的值不能比0大过设定的误差限(1E-4)，或者两次得到的x的差不超过误差限，也设定了最大迭代次数3000，超过迭代次数，则停止并报错。

```matlab
syms x
f1=x^3-x-1;
df1=diff(f1,x);
df1_=inline(df1);
f1_=inline(f1);
f2=-x^3+5*x;
f2_=inline(f2);
df2=diff(f2,x);
df2_=inline(df2);
```

### 第1个函数

```matlab
dampedNewton(f1_,df1_,0.6)
fprintf("准确解:%d",fzero(f1_,0.6))
```

得到的近似解是1.324720，fzero的结果为1.3247，前5位有效数字是一样的

再来考虑不带阻尼的，看看迭代步数有没有区别

```matlab
[res,count]=dampedNewton(f1_,df1_,0.6,false)
```

同样得到5位有效数字。但迭代步数更多，前者是4次，后者是11次，为什么出现这个差异？因为阻尼法保证了abs(f(x))是单减的，这是牛顿法无法保证的，因此它能加速判停条件abs(f(x))<ep。

### 第2个函数

```matlab
dampedNewton(f2_,df2_,1.35)
fzero(f2_,1.35)
```

同样得到了与fzero相同的5位有效数字

```matlab
[res,count]=dampedNewton(f2_,df2_,1.35)
```

相比朴素的牛顿法，迭代次数少了5次

```matlab
dampedNewton(inline(4*x^4-6*x^2-11/4),inline(diff(4*x^4-6*x^2-11/4)),0.5)
```

不过上面这个例子说明，阻尼牛顿法也未能解决牛顿法不收敛的问题，这里的函数是$4x^4-6*x^2-11/4$，起点是0.5，在0.5到-0.5之间反复摇摆

由于fzerotx需要给定一个区间`[a,b]`，而且f(a)与f(b)异号，因此先了解一下besselj零点的情况，如图：

```matlab
X = 0:0.1:10*pi;
J = besselj(0,X);
plot(X,J)
axis([0 max(X)+1 -1 1])
grid on
```

搜了搜，然后知道besselj(0,x)的零点和正弦差不多，而且逐渐具有周期性，因此如下：

```matlab
bessj0=inline('besselj(0,x)');
for n = 1:10
    z(n) = fzerotx(bessj0,[(n-1) n]*pi);
end
x = 0:pi/50:10*pi;
y = bessj0(x);
plot(z,zeros(1,10),'o',x,y,'-')
line([0 10*pi],[0 0],'color','black')
axis([0 10*pi -0.5 1.0])
```
